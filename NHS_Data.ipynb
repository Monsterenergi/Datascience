```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv ('.\\data.csv')
print(df)
```

```python
total_type1_attendances = df['Number of A&E attendances Type 1']
total_type1_attendances_4hr = df['Number of attendances over 4hrs Type 1']
```

```python
percentage = total_type1_attendances_4hr/total_type1_attendances*100
```

```python
# Filter out orgs without 4 hr+ attendances
percentage_4hr = percentage[percentage.notnull()]
```

```python
df[percentage.notnull()]
```

```python
plt.bar(df['Org name'][percentage.notnull()], percentage_4hr)
plt.xlabel('Organisations')
plt.ylabel('Percentage of Attendances over 4 hrs')
```

```python
# Extract 10 highest percentages of 4hr+ attendances
percentage.nlargest(10)
```

```python
df['Org name'][percentage.nlargest(10).index]
```

```python
plt.bar(df['Org name'][percentage.nlargest(10).index], percentage.nlargest(10))
plt.xlabel('Organisations')
plt.xticks(rotation = 90)
plt.ylabel('Percentage of Attendances over 4 hrs')
```
