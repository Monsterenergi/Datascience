```
import numpy as np
import matplotlib.pyplot as plt #the scikitlearn doesnt include an apriori method 
import pandas as pd
```

## Data Preprocessing

```
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None) #in this database it just says the names of items thats why the header is set to none
transactions = []  #the dataset is not in the ideal layout so it has to be remade and make a column such that it is in the correct framework tyhat cvan function with the apriori method
for i in range(0, 7501): #this is all the values 
  transactions.append([str(dataset.values[i,j]) for j in range(0, 20)]) #the j is the number of categories 
```

## Training the Apriori model on the dataset

```
from apyori import apriori #the minimum suypport is the number of times the event will occur in the total time period of the data and what is done is selecting the value orrepetition of a particular category and divide by the total number of data[points in the dataset 
#confidence is how accurate you want the model so you try a higher confidence and work down and tune the paramater or tune it to what you want out of the dataset 
#a good lift is generally about 3 (this is arbitrary and not a rule of thumb)
#the min lenght and max length is the number of items that you want to associate with e.g what 2 items are bought together or it would be 2 and 3 if you want buy 2 get 3
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

## Visualising the results

### Displaying the first results coming directly from the output of the apriori function

```
results = list(rules)
```

```
results
```

### Putting the results well organised into a Pandas DataFrame

```
def inspect(results): ####how is this table builts??????
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
