# Convolutional Neural Network

### Importing the libraries

```
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```
tf.__version__
```

## Part 1 - Data Preprocessing

### Preprocessing the Training set

```
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True) #rescale is feature scaling 

#but why is 255 the normalisation factor?? 

training_set = train_datagen.flow_from_directory('dataset/training_set',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'binary') #this block connects the code to the data 

#the outcome is binary either cat or dog 
#transformations avoid overfitting
#the transformations applied are geometric transformations to the images 
#augment the images 
#shrink,flip,rotate
#will create batches realtime, either real or the augmented imagers 
```

### Preprocessing the Test set

```
test_datagen = ImageDataGenerator(rescale = 1./255)
test_set = test_datagen.flow_from_directory('dataset/test_set',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'binary')



#the test data is not augmented but it is scaled i.e. feature scaling so can compare to the traioimg data 
#everything remains the same 
```

## Part 2 - Building the CNN

### Initialising the CNN

```
cnn = tf.keras.models.Sequential() #this like thr ANN creates thge NN 
```

### Step 1 - Convolution

```
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64, 64, 3]))
#THis inputs the layer of the nn 
#all the layers are part of the layers class in the keras module 
#filters are the number of features that we are trying to identify of the images 
#the dfeature detecotr is an nxn matrix so it is features that we are looking for 
#kernel is a 3x3 ?
# we want to use a rectifier activation function??? why is the rectifier acticvation function better?? 
#use a relu layer 
# the input shape is 64x64x3 the 3 is from the colour rgb if it was black and white it would be 1 
#its 64 x64 cos the image was resized 

#can modify filters and kernels these are hyperparamaters 
```

### Step 2 - Pooling

```
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

### Adding a second convolutional layer

```
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))


#exact same as first layer 
```

### Step 3 - Flattening

```
cnn.add(tf.keras.layers.Flatten())
```

### Step 4 - Full Connection

```
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

### Step 5 - Output Layer

```
cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

## Part 3 - Training the CNN

### Compiling the CNN

```
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy']) # exactly the same as ann 
```

### Training the CNN on the Training set and evaluating it on the Test set

```
cnn.fit(x = training_set, validation_data = test_set, epochs = 25)
```

## Part 4 - Making a single prediction

```
import numpy as np
from keras.preprocessing import image
test_image = image.load_img('dataset/single_prediction/cat_or_dog_1.jpg', target_size = (64, 64))
test_image = image.img_to_array(test_image) #predict method expects array 
test_image = np.expand_dims(test_image, axis = 0)
result = cnn.predict(test_image)
training_set.class_indices
if result[0][0] == 1:
  prediction = 'dog'
else:
  prediction = 'cat'
```
